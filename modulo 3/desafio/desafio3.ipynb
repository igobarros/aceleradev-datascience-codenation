```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('desafio1.csv')
```

```python
df.sample(5)
```

```python
df.describe().T
```

```python
df['estado_residencia'].value_counts()
```

```python
group_result =  df.groupby('estado_residencia')['pontuacao_credito'].agg([
    'mean'
    , 'median'
    , pd.Series.mode
    , 'std'
]).T
```

```python
group_result.index = ['media', 'mediana', 'moda', 'desvio_padrao']
```

```python
group_result = group_result.apply(lambda x: np.round(x, 2))
```

```python
group_result.to_json('submission.json')
```
